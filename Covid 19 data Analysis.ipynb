# Covid 19 Data Analysis notebook

## Import the modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
print('modules are imported.')
```

## Import Covid 19 Dataset

```python
covid_dataset_csv = pd.read_csv('Datasets/covid19_Confirmed_dataset.csv')
covid_dataset_csv.head(10)
```

# Import Total Death Dataset

```python
death_dataset_csv = pd.read_csv('Datasets/time_series_covid19_deaths_global.csv')
death_dataset_csv.head(10)
```

## Delete and marge

```python
death_dataset_csv.drop(['Lat','Long'], axis=1,inplace=True)
death_dataset_marged = death_dataset_csv.groupby('Country/Region').sum()
death_dataset_marged.head(15)
```

# Import total recovare Dataset

```python
recover_dataset_csv = pd.read_csv('Datasets/time_series_covid19_recovered_global.csv')
recover_dataset_csv.head(10)
```

## Delete and marge

```python
recover_dataset_csv.drop(['Lat','Long'], axis=1,inplace=True)
recover_dataset_marged = recover_dataset_csv.groupby('Country/Region').sum()
recover_dataset_marged.head(15)
```

## Find maximum infaction rate in a country

```python
covid_dataset_marged.loc["Bangladesh"].diff().max()
```

```python
covid_dataset_marged.loc["India"].diff().max()
```

```python
covid_dataset_marged.loc["Pakistan"].diff().max()
```

## Find max infaction rate for all country

```python
countries = list(covid_dataset_marged.index)
max_infaction_rate = []

for c in countries:
    max_infaction_rate.append(covid_dataset_marged.loc[c].diff().max())

covid_dataset_marged["max_infaction_rate"] = max_infaction_rate
```

```python
covid_dataset_marged.head(15)
```

## Create new Dataframe for needed column

```python
covid_data = pd.DataFrame(covid_dataset_marged['max_infaction_rate'])
```

```python
covid_data.head()
```
